## Loops ##

```python
#for loop

for i in range(5):
    print(i)
```

```python
for i in range (2,11,2):
    print(i)
```

```python
marks = [70,80,90]
for mark in marks:
    print(mark)
```

```python
#while loop
count = 1
while count <= 5:
    print(count)
    count+= 1
```

```python
n = 0
while True:
    if n==3:
        break
        n+=1
        print("n = ",n)
```

## Conditional Statement ##

```python
age = 18
if age >= 18:
    print("Adult")
```

```python
age = 15
if age >= 18:
    print('Adult')
else:
    print("minor")
```

```python
#if-elif (grading)
marks = 75
if marks >= 90:
    print("A")
elif marks >= 80:
    print('B')
elif marks >=70:
    print("C")
else:
    print("Fail")
```

```python
def stats(nums):
    return min(nums),max(nums),sum(nums)/len(nums)
```

```python
low,high,avg = stats([10,20,30,40])
```

```python
print(low,high,avg)
```
